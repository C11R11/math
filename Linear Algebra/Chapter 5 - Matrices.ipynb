```python
#Frobenius distance between matrices
def f_dist(A,B):
    subs = np.sum(A - B)**2
    f_dist = np.sqrt(np.sum(A - B)**2)
    return f_dist

m = 7
n = 7

A = np.random.randn(m, n)
B = np.random.randn(m, n)

scalar = 1
d = f_dist(A*scalar,B*scalar)
print(d)
i = 0

while d > 1:
    scalar *= 0.9
    d = f_dist(A*scalar,B*scalar)
    print(d)
    i += 1

print("scalar->", scalar)
print("iterations->", i)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(precision=3)
m = 10
A = np.random.randn(m,m)
norm = np.linalg.norm(A, 'fro')
norm

size = 30
shift = np.linspace(0,1,size)

shifted_norm = np.zeros(size)
correlation = np.zeros(size)
distance = np.zeros(size)

for s in range(len(shift)):
    # Matrix shift
    B = A + np.eye(m)*(norm*shift[s])
    # Matrix shifted norm
    b_norm = np.linalg.norm(B, 'fro')
    b_norm, norm
    
    #Norm difference
    norm_dif = (1 - norm/b_norm)*100
    shifted_norm[s] = norm_dif
    #print("Difference A->B", norm_dif)
    
    #Correlation
    f_A = A.flatten()
    f_B = B.flatten()
    r = np.corrcoef(f_A,f_B)[0,1]
    correlation[s] = r
    
    #print(f_A, f_B, "pearson->", r)
    
    # Matrices Distance frobenius
    def f_dist(A,B):
        subs = np.sum(A - B)**2
        f_dist = np.sqrt(np.sum(A - B)**2)
        return f_dist
    
    dist = f_dist(A,B)
    distance[s] = dist
    #print("distance A->B", dist)
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10,3))
#fig.suptitle('Horizontally stacked subplots')
ax1.plot(shifted_norm,label='shifted_norm')
ax1.grid()

ax2.plot(correlation,label='correlation')
ax2.grid()

ax3.plot(distance,label='distance')
ax3.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def GetRankedMatrix(m,r):
    A = np.random.randn(m,r)
    B = np.random.randn(r,m)
    C = A@B 
    return C

m = 20
therange = range(2,20)

sum_vals = np.zeros([m,m])

mult_vals = np.zeros([m,m])

RankInfo = np.zeros((len(therange),len(therange),2))

for i in range(len(therange)):
    for j in range(len(therange)):
        #A = GetRankedMatrix(m,therange[i])
        #B = GetRankedMatrix(m,therange[j])
        sum = np.linalg.matrix_rank(GetRankedMatrix(m,therange[i]) + GetRankedMatrix(m,therange[j]))
        mult = np.linalg.matrix_rank(GetRankedMatrix(m,therange[i]) @ GetRankedMatrix(m,therange[j]))
        RankInfo[i,j,0] = sum
        RankInfo[i,j,1] = mult

print(RankInfo[:,:,0])
print(RankInfo[:,:,1])

print(np.corrcoef(RankInfo[:,:,0].flatten(), RankInfo[:,:,1].flatten()))


plt.imshow(RankInfo[:,:,0], cmap='gray',vmin = 0, vmax = 14, origin='lower')
plt.colorbar(fraction=.045)
plt.tight_layout()
plt.show()

plt.imshow(RankInfo[:,:,1], cmap='gray', vmin=np.min(therange),vmax=np.max(therange),origin='lower',
                    extent=(therange[0],therange[-1],therange[0],therange[-1]))
plt.colorbar(fraction=.045)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

m_sizes = np.arange(3,31)

dets = np.zeros((100, len(m_sizes)))

for m in range(len(m_sizes)):
    for i in range(100):
        #The matrix
        A = np.random.randn(m_sizes[m],m_sizes[m])
        
        #print(A.shape, np.linalg.matrix_rank(A), np.linalg.det(A))
        
        #Reducing the rank
        A[:,m-1] = A[:,m-2] * 2
        
        #print("(reduced rank, det) ->", np.linalg.matrix_rank(A), np.linalg.det(A))
        
        # Determinant
        dets[i][m] = np.linalg.det(A)         

print(dets.shape, dets)

'''plt.imshow(dets, cmap='seismic',vmin = 0, vmax = 1, origin='lower')
plt.colorbar(fraction=.045)
plt.tight_layout()
plt.show()'''

aver = np.zeros(len(m_sizes))

for m in range(len(m_sizes)):
    aver[m] = np.mean(dets[:,m])

print("--->", np.mean(dets) - aver)

print(m_sizes**2, (m_sizes**2)[0].shape)
print(aver, aver[0].shape)

print(np.log(aver))

plt.plot(m_sizes**2, np.mean(dets, axis = 0))
plt.show()
```

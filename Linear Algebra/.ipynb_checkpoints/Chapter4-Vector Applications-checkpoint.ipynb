```python
# Solucion libro 
import numpy as np

# the function
def corrAndCosine(x,y):

  # compute cosine similarity
  num = np.dot(x,y) # numerator
  den = np.linalg.norm(x) * np.linalg.norm(y) # denominator
  cos = num / den

  # compute correlation (similar to above but mean-centered!)
  xm  = x-np.mean(x)
  ym  = y-np.mean(y)
  num = np.dot(xm,ym) # numerator
  den = np.linalg.norm(xm) * np.linalg.norm(ym) # denominator
  cor = num / den

  return cor,cos


# test it
a = np.random.randn(15)
b = np.random.randn(15)

# compute the correlation and cosine
r,c = corrAndCosine(a,b)

# confirm that the correlation matches with np.corrcoef
print(r,c)
print(np.corrcoef(a,b)[0,1])
```

```python
import numpy as np
import matplotlib.pyplot as plt

def coef(v1,v2):
    mean_center_v1 = v1 - np.mean(v1)
    mean_center_v2 = v2 - np.mean(v2)
    
    pearson_alg = np.dot(mean_center_v1,mean_center_v2)/(np.linalg.norm(mean_center_v1) * np.linalg.norm(mean_center_v2))
    pearson_geom = np.dot(v1,v2)/(np.linalg.norm(v1) * np.linalg.norm(v2))
    return pearson_alg, pearson_geom
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Kernel edge vector
kedge = np.array([-1, 2])
kedge

v1 = np.array([1,10])

ts1 = np.zeros(30)
#ts1 = ts1 -1
ts1[10:20] = 1
print(ts1)

plt.plot(kedge, '--s')
plt.show()

plt.plot(ts1, '-o')
plt.axis((0,30, -1.1, 1.1))
plt.show()

print(np.dot(kedge,v1))

_,axs = plt.subplots(1,2,figsize=(12,4))
axs[0].plot(kedge,'ks-')
axs[0].set_title('Kernel')
axs[0].set_xlim([-15,15])

axs[1].plot(ts1,'ks-')
axs[1].set_title('Time series signal')
plt.show()

dot = np.zeros(30)

print(ts1.size)

for idx, item in enumerate(ts1):
    if idx < ts1.size -1 :
        dot[idx] = (np.dot(kedge, np.array([ts1[idx], ts1[idx + 1]])))

print(dot.size, ts1.size)
plt.plot(dot, '-o')
plt.plot(ts1, '-o')
plt.show()

print(dot)
```

## K-Means Clustering

```python

## Create data
nPerClust = 50

# blur around centroid (std units)
blur = 1

# XY centroid locations
A = [  1, 1 ]
B = [ -3, 1 ]
C = [  3, 3 ]

# generate data
a = [ A[0]+np.random.randn(nPerClust)*blur , A[1]+np.random.randn(nPerClust)*blur ]
b = [ B[0]+np.random.randn(nPerClust)*blur , B[1]+np.random.randn(nPerClust)*blur ]
c = [ C[0]+np.random.randn(nPerClust)*blur , C[1]+np.random.randn(nPerClust)*blur ]

print(len(a), len(a[0]), len(a[1]))
print(len(b), len(b[0]), len(b[1]))
print(len(c), len(c[0]), len(c[1]))

# concatanate into a matrix
data_c = np.concatenate((a,b,c),axis=1)
print(data_c.shape)
data = np.transpose(data_c)

print(data.shape)

# plot data
plt.plot(data[:,0],data[:,1],'ko',markerfacecolor='w')
plt.plot(A[0], A[1],'v')
plt.plot(B[0], B[1],'v')
plt.plot(C[0], C[1],'v')
plt.title('Raw (preclustered) data')
plt.xticks([])
plt.yticks([])
plt.grid()
plt.show()
```

```python
## initialize random cluster centroids
k = 4 # extract three clusters

# random cluster centers (randomly sampled data points)
ridx = np.random.choice(range(len(data)),k,replace=False)
print(ridx)
centroids = data[ridx,:]
print(centroids)

# setup the figure
fig,axs = plt.subplots(3,3,figsize=(6,6))
axs = axs.flatten()
lineColors = [ [0,0,0],[.4,.4,.4],[.8,.8,.8] ]#'rbm'


# plot data with initial random cluster centroids
axs[0].plot(data[:,0],data[:,1],'ko',markerfacecolor='w')
axs[0].plot(centroids[:,0],centroids[:,1],'ko')
axs[0].set_title('Iteration 0')
axs[0].set_xticks([])
axs[0].set_yticks([])



# loop over iterations
for iteri in range(5):
    
  # step 1: compute distances
  dists = np.zeros((data.shape[0],k))
  for ci in range(k):
    dists[:,ci] = np.sum((data-centroids[ci,:])**2,axis=1)
        
  # step 2: assign to group based on minimum distance
  groupidx = np.argmin(dists,axis=1)
    
  # step 3: recompute centers
  for ki in range(k):
    centroids[ki,:] = [ np.mean(data[groupidx==ki,0]), np.mean(data[groupidx==ki,1]) ]
  

  # plot data points
  for i in range(len(data)):
    axs[iteri+1].plot([ data[i,0],centroids[groupidx[i],0] ],[ data[i,1],centroids[groupidx[i],1] ])#,color=lineColors[groupidx[i]])
  axs[iteri+1].plot(centroids[:,0],centroids[:,1],'ko')
  axs[iteri+1].set_title(f'Iteration {iteri+1}')
  axs[iteri+1].set_xticks([])
  axs[iteri+1].set_yticks([])

plt.show()
```

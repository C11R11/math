```python
import numpy as np

a = np.array([-4, -2, 0, 2, 4])
b = np.array([ -12, -6, 0,6,9])
```

## Variance


## Covarianza


```python
cov = np.cov(a,b)
cov_h = np.sum((a - np.mean(a))*(b - np.mean(b)))/(len(a)-1)

cov, cov_h
```

## Standard deviation



```python
dev = np.std(a)
dev_h = np.sqrt(np.sum(((a - np.mean(a))**2)/len(a)))
dev, dev_h
```

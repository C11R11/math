## Covariance matrix

```python
import numpy as np
import matplotlib.pyplot as plt

m = 10
A = np.random.randint(0,20, size=(m,m))

print("Matrix->", A)

# Check covariance
# Taking the first two rows from the dataset and computing the covariance

f1 = A[:,0]
var_f1 = np.sum((f1-np.mean(f1))**2)/(len(f1))
print(f1, f1.shape, var_f1, np.var(f1))

f2 = A[:,1]
var_f2 = np.sum((f2-np.mean(f2))**2)/(len(f2))
print(f2, f2.shape, var_f2, np.var(f2))

cov_f1_f2 = np.sum((f1 - np.mean(f1))*(f2 - np.mean(f2)))/(len(f1)-1)


print("cov_f1_f2->", cov_f1_f2, np.cov(f1, f2), np.var(f1), np.var(f2))


# compute the mean of each data feature
datamean = np.mean(A,axis=0)

# mean-center the data using broadcasting
dataMatM = A - datamean

# Now to compute the covariance matrix
covMat = dataMatM.T @ dataMatM  # data matrix time
covMat /= (m-1)

# When you have the variables to analyze in columns and want to use the covariance
# It's needed to transpose the matrix or use the rowvar option
#covNP = np.cov(A, rowvar=False)
covNP = np.cov(A.T)

covMat[0,1], covMat[1,0] , covNP[0,1], covNP[1,0]
```
